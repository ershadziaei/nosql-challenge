### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': 20}
results = establishments.find(query)
print(results)

```

```python
# Use count_documents to display the number of documents in the result
print("Number of documents in the result:", establishments.count_documents(query))

# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)

# Display the number of rows in the DataFrame

print(f"There are a total of {len(df)} rows in the DataFrame\n")

# Display the first 10 rows of the DataFrame
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
new_query = {'LocalAuthorityName': {'$regex': 'London'}, 'RatingValue': {'$gte': '4'}}
new_results = establishments.find(new_query)

# Use count_documents to display the number of documents in the result

print(f"\nThere are a total of {establishments.count_documents(new_query)} documents in the results for London.\n")

# Display the first document in the results using pprint
pprint(new_results[0])
```

```python
# Convert the result to a Pandas DataFrame
new_df = pd.DataFrame(new_results)

# Display the number of rows in the DataFrame
print(f"There are a total of {len(new_df)} rows in the DataFrame\n")

# Display the first 10 rows of the DataFrame
new_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = establishments.find_one({'BusinessName': 'Penang Flavours'}, {'geocode.latitude'})['geocode']['latitude']
longitude = establishments.find_one({'BusinessName': 'Penang Flavours'}, {'geocode.longitude'})['geocode']['longitude']

third_query = {'geocode.latitude': {'$gte': (latitude - degree_search), '$lte': (latitude + degree_search)}, 
        'geocode.longitude': {'$gte': (longitude - degree_search), '$lte': (longitude + degree_search)},
        'RatingValue': '5'}

sort =  [('scores.Hygiene', 1)]

# Print the results
results3 = list(establishments.find(third_query).sort(sort).limit(5))
pprint(results3)
```

```python
# Convert result to Pandas DataFrame
dataframe3 = pd.DataFrame(results3)
dataframe3
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
match = {'$match': {'scores.Hygiene': 0}}

# 2. Groups the matches by Local Authority
group = {'$group': {'_id': {'LocalAuthorityName': '$LocalAuthorityName'}, 'count': {'$sum': 1}}}

# 3. Sorts the matches from highest to lowest
sortt = {'$sort': {'count': -1}}

pipeline = [match, group, sortt]
pip_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result

print(f"There are a total of {len(pip_results)} Local Authority areas with establishments that have a Hygiene Score of 0\n")

# Print the first 10 results
pprint(pip_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
pip_result_df =  pd.json_normalize(pip_results)
pip_result_df
```

```python
# Display the number of rows in the DataFrame
print(f"There are a total of {len(pip_result_df)} Local Authority areas.\n")
```
